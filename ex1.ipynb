```python
print("==예제1==")
s = "안녕하세요"
#   [0, 1, 2, 3, 4]

print("==s길이 출력==")
print(len(s))

# s[start:end(-1):step]
print("s를 이루는 문자 중에서 앞에 거 2개 출력")
print(s[0:2])

print("s를 이루는 문장 중에서 문자 하나 건너 뛰고 1개 출력")
print(s[1:2])

print("s를 이루는 문자 중에서 문자 4개 출력, 첫 인덱스 생략")
print(s[:4])

print("s를 이루는 문장 중에서 앞에 문자 3개 건너뛰고, 나머지 전부 출력, len")
print(s[3:len(s)])

print("s를 이루는 문장 중에서 앞에 문자 3개 건너뛰고, 뒤 인덱스 생략")
print(s[3:])

```

```python
print("==예제2==")
s = "안녕하세요"

print("s를 이루는 문자를 순서대로 하나씩 출력, while")
i = 0
while i < len(s):
    print(s[i])
    i += 1  
    

print("s를 이루는 문자를  역순으로 하나씩 출력, while")
i = len(s) - 1
while i >= 0:
    print(s[i])
    i -= 1

print("s를 이루는 문자를  역순으로 하나씩 출력, for")
for c in s:
    print(c)
```

```python
print("==예제3==")
s= "a안b녕c하d세e요f"
print("s를 이루는 문자 중에서 한글만 출력, in 방법3")
for c in s:
    if c not in ["a", "b", "c", "d", "e", "f"]:
        print(c)


s= "a안b녕c하d세e요f"
print("s를 이루는 문자 중에서 한글만 출력, 방법1")
print(s[1] + s[3] + s[5] + s[7] + s[9])

print("s를 이루는 문자 중에서 한글만 출력, for, if, and, 방법2")
for c in s:
    if c.isalpha() and ord(c) > 127:
        print(c, end="")
```
