```python
# 클래스 ->객체를 표현하기 위한 문법의 일종, 프로그래밍으로 객체를 만들 때 사용하는 것

# 객체지향언어-파이썬 

# 게임캐릭터 -전사, 마법사, 궁수, 자동차, 비행기 -> object: 현실, 가상, 개념

class Person: # 클래스는 (객체를 만들기 위한) 설계도이다. 붕어빵 틀
    def hello(self):
        print("hi")

jin = Person() # jin이라는 객체가 Person이라는 클래스의 인스턴스가 된다. 설계도를 보고 만든 진짜 사람이 된다. jin은 Person의 인스턴스이다. 인스턴스가 실제 객체이다. 붕어빵
jin.hello() # jin이라는 객체가 hello라는 메소드를 사용한다. 붕어빵을 찍어낸다.

```

```python
class Person:
    def hello(self):
        print("hello")

    def hi(self):
        self.hello() # self는 자기자신을 가리키는 예약어이다. 자기자신의 hello메소드를 호출한다. 
        # self 는 리모콘, self.hello()는 리모콘으로 hello메소드를 호출한다. self는 객체 자신을 가리키는 예약어이다. 자기자신의 hello메소드를 호출한다.   

jin = Person()
jin.hi()
```
