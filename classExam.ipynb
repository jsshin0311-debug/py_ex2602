```python
class Person:
    def hello(self):
        print("hello")

    def hi(self):
        self.hello() # self는 자기자신을 가리키는 예약어이다. 자기자신의 hello메소드를 호출한다. 
        # self 는 리모콘, self.hello()는 리모콘으로 hello메소드를 호출한다. self는 객체 자신을 가리키는 예약어이다. 자기자신의 hello메소드를 호출한다.   

jin = Person()
jin.hi()
```

```python
class Person():
    def __init__(self):  # __init__는 생성자이다. 메소드
        self.hello = "안녕하세요" # 인스턴스 속성
        #인스턴스 속성

    def hi(self):
        print(self.hello) #리모콘을 통해 이 속성을 가져온다.

jin = Person() # 이 때 __init__ 메서드는 이미 실행됐다.
jin.hi()
```

```python
class Person():
    def __init__(self, name, age, hobby):  
        self.hello = "안녕하세요" # self.hello는 객체의 속성이다. 객체의 속성은 객체가 가지고 있는 데이터이다. self.hello는 jin이라는 객체가 가지고 있는 hello라는 속성이다. jin.hello는 jin이라는 객체가 가지고 있는 hello라는 속성의 값을 출력한다.
        self.name = name
        self.age = age
        self.hobby = hobby
        

    def introduce(self):
        print("{}. 저는 {}입니다." .format(self.hello, self.name))

jin = Person('진', 20, '코딩')
jin.introduce()

print('이름:', jin.name)
print('나이:', jin.age)
print('취미:', jin.hobby)
```

```python
# 클래스 외부에서는 접근 불가, 클래스 내부에서만 사용 가능한 private attribute(속성) 만들기
# 언더바 두개 : 비공개 속성, 클래스 내부에서만 접근 가능, 클래스 외부에서는 접근 불가
class Person():
    def __init__(self, name, age, hobby, weight):  
        self.hello = "안녕하세요"
        self.name = name
        self.age = age
        self.hobby = hobby
        self.__weight = weight 

jin = Person('진', 22, '운동', 78)
jin.__weight -= 20  #jin.__weight = __weight - 20의 줄임말이다.

print(jin.__weight) #클래스 밖에서는 __~~에 접근이 안된다.

#결과값은 오류가 난다.
```

```python
class Person():
    def __init__(self, name, age, hobby, weight):  
        self.hello = "안녕하세요"
        self.name = name
        self.age = age
        self.hobby = hobby
        self.__weight = weight  #private attribute

#비공개 속성에 우회해서 접근하기 aka.VPN 
#몸무게 감소 메서드
    def diet(self, loss_weight): 
        self.__weight -= loss_weight
        print("몸무게 :", self.__weight)

jin = Person('진', 22, '운동', 78)
jin.diet(10)

```

```python
#26-02-26
#저번시간 복습
#저번시간 까지는 인스턴스 속성에 대해 배움
#인스턴스 속성 => __init__

class Person():
    def __init__(self, name, age, hobby): 
        self.hello = "안녕하세요"
        self.name = name
        self.age = age
        self.hobby = hobby




#속성
#1. 인스턴스 속성(attribute)
#2. 클래스 속성
```

```python
#클래스 속성

class Person():
    folder = [] #클래스 속성 -> 클래스에 속한다. (공공재) 모든 인스턴스에서 접근이 가능.

    def add_files(self, file):
        Person.folder.append(file) #클래스 이름으로 클래스 속성에 접근할 수 있다.

jin = Person()
jin.add_files("txt")

paul = Person()
paul.add_files("img")

print(jin.folder)
print(paul.folder)


```

```python
# 인스턴스 속성

class Person():
    def __init__(self):
        self.folder = [] #인스턴스 속성
    
    def add_files(self, file):
        self.folder.append(file) 


jin = Person()
jin.add_files('txt')

paul = Person()
paul.add_files('img')

print(jin.folder)
print(paul.folder)
        
```
