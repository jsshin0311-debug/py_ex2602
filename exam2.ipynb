```python
#파이썬유니코드

#1. 문자열은 컴퓨터 내부에서 숫자(유니코드)로 저장된다.
#-내부적으로는 숫자로 저장되지만, 우리는 문자로 볼 수 있다.
#-그 숫자 체계를 유니코드라고 한다.

text = "파이썬"

print(text) #문자열로 출력
print(len(text)) #문자열의 길이(문자 수)를 출력

```

```python
#2. 문자 <-> 숫자 변환

print(ord("가")) #문자 'A'의 유니코드 숫자값을 출력
print(chr(44032)) #유니코드 숫자값 44032에 해당하는 문자를 출력
```

```python
#3. 문자열을 바이트로 변환 (encode)
# - encode : 문자열을 바이트(byte)로 변환
#파일 저장이나 네트워크 전송 시 사용됨

encoded = text.encode("utf-8")
print(encode) #b'\xec\x95\x88\xeb\x85\x95'
print(type(encoded)) #<class 'bytes'>
```

```python
# 4. 바이트를 다시 문자열로 변환(decode)
# - decode : 바이트를 문자열로 변환
decoded = encoded.decode('utf-8') 
print(decoded) 
print(type(decoded)) #<class 'str'>
```
