```python
#openpyxl => 파이썬에서 엑셀파일을 읽고 쓰기 위한 라이브러리

# openpyxl 불러오기 / * : 모든 기능을 불러옴
import openpyxl
#from openpyxl import *

print(openpyxl.__version__)

#====엑셀 파일 만들기, 기본 시트 선택, 셀에 값 넣기 ====
#새로운 엑셀 파일 만들기
#새 엑셀 파일을 만들겠다는 의미
w1 = openpyxl.Workbook()

#만들어진 엑셀 파일에서 기본 시트를 선택
ws = w1.active

#지금 만든 엑셀 파일을 실제 파일로 저장
#엑셀 파일이름을 test.xlsx로 지정해서 저장
#w1.save("test.xlsx")

#기존에 저장된 엑셀 파일을 다시 열어서 수정할 수 있게 함
w2 = openpyxl.load_workbook("test.xlsx")

#열어온 엑셀 파일에서 첫 번째 시트 선택
ws2 = w2.active

#셀에 값 입력하기
#A열 1행 셀에 hello라는 문자열 입력
ws2["A1"] = 'hello'

#변경사항 저장
w2.save("test.xlsx")

#C열 3행 셀에 bye라는 문자열 입력
ws2["C3"] = 'bye'

#변경사항 저장
w2.save("test.xlsx")

a = 'python'
#D열 5행부터 F열 8행까지의 셀에 a라는 문자열 입력
for row in ws2["D5" : "F8"]: #행 단위로 반복
    for cell in row: #각 행 안의 셀들을 하나씩 꺼내서
        cell.value = a #셀에 python이라는 문자열 입력
w2.save("test.xlsx") #변경사항 저장


num = 2
#A열 1행부터 C열 3행까지 범위 선택 
for row in ws2["A1" : "C3"]:
    for cell in row:
        cell.value = num #셀에 2라는 숫자 입력
        num+= 2 #2씩 증가
w2.save("test.xlsx") #변경사항 저장
```

```python

from openpyxl import *

#=====딕셔너리로 값 생성 후 엑셀에 넣기 =====

item1 = {
    '품목': '바나나',
    '가격': '30000',
    '수량': 10,
    '날짜': '2026-02-20'
}

item2 = {
    '품목': '사과', 
    '가격': '50000',
    '수량': 5,
    '날짜': '2026-02-21'
}

item3 = {
    '품목': '딸기', 
    '가격': '20000',
    '수량': 20,
    '날짜': '2026-02-22'
}

item4 = {
    '품목': '포도',
    '가격': '40000',
    '수량': 15,
    '날짜': '2026-02-23'
}

wb = Workbook()
ws = wb.active


ws['A1'] = '품목'
ws['B1'] = '가격'     
ws['C1'] = '수량'
ws['D1'] = '날짜'

items = [item1, item2, item3, item4] 

i = 2
for item in items:
    ws['A' + str(i)] = item['품목']
    ws['B' + str(i)] = item['가격']
    ws['C' + str(i)] = item['수량']
    ws['D' + str(i)] = item['날짜']
    i += 1
wb.save("test2.xlsx")
```
