```python
import matplotlib.pyplot as plt

plt.style.use('default') #그래프 스타일을 기본으로 설정

#y값만 전달, x축은 자동으로 0,1,2,3,...로 설정
#plt.plot([2,4,6,8], [0,1,2,3]) #x값이 2,4,6,8이고 y값이 0,1,2,3인 그래프를 그린다.
plt.plot([2,4,6,8]) #y값이 2,4,6,8인 그래프를 그린다
plt.show() #그래프를 화면에 보여준다
```

```python
fig = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] #막대의 위치(좌표기준)
y = [1, 2] #막대의 높이(값 기준)

#x축 라벨 이름 (tick 이름)
labels = ['english', 'math']

#각 막대에 적용할 색상
colors = ['red', 'blue']    

a.bar(x, y, color=colors, tick_label=labels)
plt.show()  
```

```python
fig = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] #막대의 위치(좌표기준)
y = [1, 2] #막대의 높이(값 기준)

#x축 라벨 이름 (tick 이름)
labels = ['english', 'math']

#각 막대에 적용할 색상
colors = ['red', 'blue']    

a.barh(x, y, color=colors, tick_label=labels) #h를 붙이면 가로로 나옴, 수평 막대 그래프
plt.show()  
```

```python
fig = plt.figure()

plt.style.use("ggplot")

a = fig.add_subplot(111)
x = [3, 4] #막대의 위치(좌표기준)
y = [1, 2] #막대의 높이(값 기준)

#x축 라벨 이름 (tick 이름)
labels = ['english', 'math']

#각 막대에 적용할 색상
colors = ['red', 'blue']    

a.bar(x, y, color=colors, tick_label=labels)

for i, v in enumerate(y):
    a.text(v+2, i+1.05, str(v)) #표준수치, 막대 위에 텍스트를 추가하는 함수, v+2는 텍스트의 x좌표, i+1.05는 텍스트의 y좌표, str(v)는 텍스트로 표시할 값



plt.show()  
```
